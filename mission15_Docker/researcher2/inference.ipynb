```python
import pandas as pd
import pickle
import os

# 모델 로드
with open('model.pkl', 'rb') as f:
    model = pickle.load(f)

# 데이터 로드
test_df = pd.read_csv('data/mission15_test.csv')
test_df.head()
```

```python
# 데이터 전처리 
test_df["Extracurricular Activities"] = test_df["Extracurricular Activities"].map({"No": 0, "Yes": 1})

# 만약 test.csv에 정답(Performance Index) 컬럼이 껴있다면 빼고 문제만 줍니다.
if 'Performance Index' in test_df.columns:
    X_test = test_df.drop('Performance Index', axis=1)
else:
    X_test = test_df
```

```python
# 추론
predictions = model.predict(X_test)

# 결과 저장 (result.csv)
result_df = pd.DataFrame({
    'Prediction': predictions
})

# index=False 를 줘야 쓸데없는 순번 열이 안 생깁니다.
output_path = "result.csv"
result_df.to_csv(output_path, index=False)

print(f"결과가 '{output_path}'에 저장되었습니다.")
```
